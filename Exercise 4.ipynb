```python
#Programs using matplotlib/plotly/bokeh/seaborn for data visualisation.

import numpy as np
import matplotlib.pyplot as plt

a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,marker='*')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,marker='p')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,marker='p',ms=20)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,marker='p',mec='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,marker='p',mfc='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Graph',loc='left')
plt.plot(a,b,marker='p',mfc='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.scatter(a,b)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.grid()
plt.plot(a,b,marker='p',mfc='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.grid(axis='x')
plt.plot(a,b,marker='p',mfc='r')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.plot(a,b,'o:g')
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.bar(a,b)
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([1,2,3,4,5])
plt.bar(a,b,width=0.1)
plt.show()
```

```python
a=np.array(["A","B","c","D"])
b=np.array([3,8,1,10])
plt.barh(a,b)
plt.show()
```

```python
a=np.array(["A","B","c","D"])
b=np.array([3,8,1,10])
plt.barh(a,b,height=0.1)
plt.show()
```

```python
a=np.array(["A","B","c","D"])
b=np.array([3,8,1,10])
plt.bar(a,b,color='red')
plt.show()
```

```python
b=np.array([35,25,25,15])
plt.pie(b)
plt.show()
```

```python
b=np.array([35,25,25,15])
a=["Apple","Orange","Grapes","Mango"]
plt.pie(b,labels=a)
plt.show()
```

```python
b=np.array([35,25,25,15])
a=["Apple","Orange","Grapes","Mango"]
c=[0.2,0,0,0]
plt.pie(b,labels=a,explode=c)
plt.show()
```

```python
b=np.array([35,25,25,15])
a=["Apple","Orange","Grapes","Mango"]
c=[0.2,0,0,0]
plt.pie(b,labels=a,explode=c,shadow=True)
plt.show()
```

```python
b=np.array([35,25,25,15])
a=["Apple","Orange","Grapes","Mango"]
c=[0.2,0,0,0]
d=["black","yellow","blue","green"]
plt.pie(b,labels=a,explode=c,colors=d)
plt.show()
```

```python
b=np.array([35,25,25,15])
a=["Apple","Orange","Grapes","Mango"]
d=["black","yellow","blue","green"]
plt.pie(b,labels=a,colors=d)
plt.legend()
plt.show()
```
